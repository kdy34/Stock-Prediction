**Importing Scikit-Learn libraries**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x = df.drop("Volume", axis = 1)
y = df["Volume"]
```

```python
print(f"Shape of x: {x.shape}")
print(f"Shape of y: {y.shape}")
```

**Splitting the data**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 15)
```

```python
print(f"Shape of x_train: {x_train.shape}")
print(f"Shape of x_test: {x_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

**Scaling the data**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

**Creating the Model**

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
print(pred)
```

**Evaluating the prediction**

```python
print(f"Mean Squared Error: {mean_squared_error(y_test, pred)}")
print(f"R2 Score: {r2_score(y_test, pred)}")
```

**Visualizing the prediction values**

```python
plt.scatter(y_test, pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test, pred, color='blue', edgecolor='k', alpha=0.6, s=100)

min_val = min(min(y_test), min(pred))
max_val = max(max(y_test), max(pred))

plt.plot([min_val, max_val], [min_val, max_val], color='green', lw=2, linestyle=':', label='Perfect Fit')

# Labels and Title
plt.xlabel("Actual Values", fontsize=10)
plt.ylabel("Predicted Values", fontsize=10)
plt.title("Actual vs Predicted Values", fontsize=12)

# Legend
plt.legend(fontsize=10)

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()

plt.show()
```
